```python
import json
import os

import matplotlib.pyplot as plt
import numpy as np
```

```python
with open('../temp/test_video_order_target.json', 'r') as f:
    target = json.load(f)
with open('../temp/test_video_order_output.json', 'r') as f:
    output = json.load(f)
with open('../temp/test_video_order_image.json', 'r') as f:
    image = json.load(f)
```

```python
len(target), len(output), target[0].keys(), output['0'].keys()
```

```python
image = np.array(image)
```

```python
pred_masks_high_res = []
for frame_idx in output:
    pred_masks_high_res.append(np.array(output[frame_idx]['pred_masks_high_res']))
pred_masks_high_res = np.array(pred_masks_high_res).squeeze(1)
```

```python
target_masks_high_res = []
for frame_tgt in target:
    target_masks_high_res.append(np.array(frame_tgt['masks']))
target_masks_high_res = np.array(target_masks_high_res)
```

```python
target_masks_high_res.shape
```

```python
B = image.shape[0]
figsize = np.array([B, 2]) * 5
fig, axs = plt.subplots(B, 2, figsize=figsize, )
obj_idx = 16  # Index of the object to visualize

for ii in range(B):
    axs[ii, 0].imshow(image[ii])
    axs[ii, 0].axis('off')
    axs[ii, 0].set_title(f'Image {ii+1}')

    axs[ii, 1].imshow(target_masks_high_res[ii, obj_idx:obj_idx+1].transpose(1, 2, 0))
    axs[ii, 1].axis('off')
    axs[ii, 1].set_title('PRED')

plt.tight_layout()
plt.show()
```
